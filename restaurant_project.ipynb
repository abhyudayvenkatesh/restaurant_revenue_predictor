```python
data=data.drop('Name',axis=1)
data
```

```python
data['Parking Availability']=data['Parking Availability'].apply(lambda x: 1 if x=='Yes' else 0)
# data['Parking Availability'].value_counts()
data
```

```python
data=data.drop('Cuisine', axis=1)
data
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(data.corr(), annot=True)
```

```python
data=data.drop('Parking Availability', axis=1)
data=data.drop('Service Quality Score', axis=1)
data=data.drop('Ambience Score', axis=1)
data=data.drop('Avg Review Length', axis=1)
data=data.drop('Number of Reviews', axis=1)
data=data.drop('Rating', axis=1)
plt.figure(figsize=(15,10))
sns.heatmap(data.corr(), annot=True)
```

```python
data.hist(figsize=(15,8))
```

```python
x=data.drop('Revenue',axis=1)
y=data['Revenue']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
x_scaled=scaler.fit_transform(x_train)
x_scaled_test=scaler.transform(x_test)

```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_scaled,y_train)
y_pred=model.predict(x_scaled_test)
model.score(x_scaled_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
model.fit(x_scaled,y_train)
y_pred=model.predict(x_scaled_test)
model.score(x_scaled_test,y_test)
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
score
```

```python
from sklearn.model_selection import GridSearchCV
model=RandomForestRegressor()

param_grid={
    'n_estimators':[100,200,300],
    'max_features':[2,4,6,8],
    'criterion':['squared_error','absolute_error']
}
grid=GridSearchCV(model,param_grid,refit=True,verbose=3,cv=5)
grid.fit(x_scaled,y_train)
```

```python
grid.best_estimator_
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor # Make sure to import the model

model=RandomForestRegressor()

param_grid={
    'n_estimators':[100,200,300],
    'max_features':[2,4,6,8],
    'criterion':['squared_error','absolute_error']
}
grid=GridSearchCV(model,param_grid,refit=True,verbose=3,cv=5)

# Capture potential errors during fitting
try:
    grid.fit(x_scaled,y_train)
    print(grid.best_estimator_)
except Exception as e:
    print(f"An error occurred during fitting: {e}")
```
